```python
text_list = full.iloc[:,5]
```

```python
id_list = full.iloc[:,1]
lab_list = full.iloc[:,0]
index_list = list(range(len(id_list))) ##create data index
```

```python
import random
random.Random(1234).shuffle(index_list) ##shuffle the dataset
```

```python
i_n_list = [id_list[i] for i in index_list[20000:30000]] ##extracting userID from shuffled data index from(20kth to 30kth)
```

```python
lab_s_list = [lab_list[j] for j in index_list[20000:30000]] ##extracting sentiment labels from shuffled data index
```

```python
import time
#time.sleep(43200)
t = i_n_list
expan = "expansions=author_id"
field = "user.fields=location"
nfid = []
dataid = []
locinfo = []
labs = []
inds = []
e = []
count = 1
for i in range(10000):
  ids = "ids=" + t[i]
  url = "https://api.twitter.com/2/tweets?{}&tweet.fields=geo&{}&{}".format(ids, expan, field)
  #print(url)
  js = connect_to_endpoint(url)
  if "errors" in js:
    nfid.append(js["errors"][0]["value"])
  elif "data" in js:
    dataid.append(js["data"][0]["id"])
    inds.append(index_list[i])
    labs.append(lab_s_list[i])
    print(js["data"][0]["text"])
  ###extracting location data from user profiles
    if "location" in js["includes"]["users"][0]:
      locinfo.append(js["includes"]["users"][0]["location"])
    else:
      locinfo.append("N/A")
  else:
    e.append(js)
  print(count)
  count = count + 1
  time.sleep(4)
print(time.localtime())
print("done")
```

```python
dfnew = pd.DataFrame(list(zip(dataid,locinfo,inds,labs)), columns = ["id", "locationInfo", "index", "labels"], dtype = str)
dfnew.to_csv("locIn20-30k.csv", index = False)
dfnew.to_csv("/content/gdrive/MyDrive/sst2/locInf20-30k.csv", index = False)  ##saving data into files for geographic tagging
print("done")
#with open("21kINDEX.txt", "w") as l:
#  l.write(",".join(str(inds)))
#print("done")
```
